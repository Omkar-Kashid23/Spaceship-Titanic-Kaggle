```python
# Extract the Group number (everything before the underscore)
df['Group'] = df['PassengerId'].str.split('_').str[0]

# Group the data and fill missing planets with known planets from the same group
df['HomePlanet'] = df.groupby('Group')['HomePlanet'].transform(lambda x: x.ffill().bfill())
```

```python
# Extract the Deck letter (the first letter before the slash)
df['Deck'] = df['Cabin'].str.split('/').str[0]

# Assign Europa to solo passengers on decks A, B, C, or T
df.loc[(df['HomePlanet'].isna()) & (df['Deck'].isin(['A', 'B', 'C', 'T'])), 'HomePlanet'] = 'Europa'

# Assign Earth to solo passengers on deck G
df.loc[(df['HomePlanet'].isna()) & (df['Deck'] == 'G'), 'HomePlanet'] = 'Earth'
```

```python
# Extract the Deck letter (the first letter before the slash)
df['Deck'] = df['Cabin'].str.split('/').str[0]

# Assign Europa to solo passengers on decks A, B, C, or T
df.loc[(df['HomePlanet'].isna()) & (df['Deck'].isin(['A', 'B', 'C', 'T'])), 'HomePlanet'] = 'Europa'

# Assign Earth to solo passengers on deck G
df.loc[(df['HomePlanet'].isna()) & (df['Deck'] == 'G'), 'HomePlanet'] = 'Earth'
```

```python
df.isnull().sum()
```

```python
# Create a list of all the luxury/spending columns
spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

# If a passenger spent ANY money, they were awake. Set CryoSleep to False.
df.loc[(df['CryoSleep'].isna()) & (df[spending_cols].sum(axis=1) > 0), 'CryoSleep'] = False

# If a passenger spent exactly 0.0 total, it's highly likely they were asleep. Set CryoSleep to True.
df.loc[(df['CryoSleep'].isna()) & (df[spending_cols].sum(axis=1) == 0), 'CryoSleep'] = True

# Fill any tiny remaining handful with the majority class (False)
df['CryoSleep'] = df['CryoSleep'].fillna(False)
```

```python
# For each spending column, if they are in CryoSleep, their missing spending is 0.0
for col in spending_cols:
    df.loc[(df[col].isna()) & (df['CryoSleep'] == True), col] = 0.0

# For the remaining awake passengers, fill their missing spending with the median (usually 0.0 anyway)
for col in spending_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Fill Destination with the mode
df['Destination'] = df['Destination'].fillna('TRAPPIST-1e')

# Fill missing Cabins with a placeholder
df['Cabin'] = df['Cabin'].fillna('U/U/U')
```

```python
# 1. Passengers on Deck G are never VIPs
df.loc[(df['VIP'].isna()) & (df['Deck'] == 'G'), 'VIP'] = False

# 2. Passengers from Earth are never VIPs
df.loc[(df['VIP'].isna()) & (df['HomePlanet'] == 'Earth'), 'VIP'] = False

# 3. For any remaining missing values, fill with False (since VIPs are extremely rare)
df['VIP'] = df['VIP'].fillna(False)
```

```python
df['Deck'] = df.groupby('Group')['Deck'].transform(lambda x: x.ffill().bfill())
df['Deck'] = df['Deck'].fillna('U')
if 'Side' in df.columns:
    df['Side'] = df['Side'].fillna('U')
```

```python
# 1. Extract the Group number
df['Group'] = df['PassengerId'].str.split('_').str[0]

# 2. Count how many times each Group number appears to get the Group Size
df['Group_Size'] = df['Group'].map(df['Group'].value_counts())

# 3. Create a feature showing if they are traveling alone (True/False)
df['Is_Solo'] = df['Group_Size'] == 1

# 4. Now that we extracted the value, drop the raw columns from your TRAINING data
# (Make sure to keep PassengerId in your test dataset for the final submission!)
X_train = df.drop(['PassengerId', 'Group'], axis=1)
```

```python
# Create a list of the spending columns again
spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

# Sum them up into a single new column
df['Total_Spending'] = df[spending_cols].sum(axis=1)

# Bonus feature: Create a True/False column if they spent absolutely nothing
df['No_Spending'] = df['Total_Spending'] == 0
```

```python
# Extract the Side (the last letter after the second slash)
# We use .str[-1] to just grab the very last character
df['Side'] = df['Cabin'].str[-1]

# Drop the original Cabin column since the model can't read strings like "B/1/P"
df = df.drop('Cabin', axis=1)

# Drop Name if you haven't already
df = df.drop('Name', axis=1)
```

```python
# Convert all True/False columns to 1 and 0
bool_cols = ['CryoSleep', 'VIP', 'Is_Solo', 'No_Spending']
for col in bool_cols:
    df[col] = df[col].astype(int)

# One-Hot Encode the text columns (HomePlanet, Destination, Deck, Side)
df = pd.get_dummies(df, columns=['HomePlanet', 'Destination', 'Deck', 'Side'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Create a list of ONLY your continuous numerical columns
num_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Total_Spending', 'Group_Size']

# 2. Initialize the scaler
scaler = StandardScaler()

# 3. Fit and transform the data, replacing the old columns with the scaled ones
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# 1. Define the "grid" of settings you want to test
# 'C' controls regularization (smaller values = stronger penalty for overfitting)
# 'solver' is the mathematical algorithm used to find the best fit
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs']
}

# 2. Initialize a fresh, blank model
base_lr = LogisticRegression(max_iter=2000) # Bumped up max_iter to give the solvers plenty of time

# 3. Set up the GridSearchCV
# cv=5 means 5-fold cross-validation (it tests itself 5 times for every combination)
grid_search = GridSearchCV(estimator=base_lr, param_grid=param_grid, cv=5, scoring='accuracy')

# 4. Start the search! (This might take a minute or two)
print("Searching for the best hyperparameters...")
grid_search.fit(x_train, y_train)

# 5. Print out the winning combination
print(f"Best Parameters Found: {grid_search.best_params_}")
print(f"Best Cross-Validation Score: {grid_search.best_score_ * 100:.2f}%")
```

```python
from sklearn.metrics import accuracy_score

# 1. Extract the absolute best model from the search
best_model = grid_search.best_estimator_

# 2. Make predictions on your unseen test data
tuned_predictions = best_model.predict(x_test)

# 3. See if the tuned model beat your original baseline!
final_score = accuracy_score(y_test, tuned_predictions)
print(f"Final Tuned Accuracy: {final_score * 100:.2f}%")
```
